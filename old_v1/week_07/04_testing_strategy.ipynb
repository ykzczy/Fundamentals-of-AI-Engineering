### What this part covers
This notebook covers **testing strategy** — writing tests that protect you from regressions and give you confidence that your pipeline works correctly.

**The minimum test plan (3 cases):**
1. **Happy path** — normal valid input → expected output
2. **Edge case** — unusual but valid input (empty columns, all-null column, single-row CSV)
3. **Failure case** — invalid input → expected error type and message

**Why test LLM code without calling the LLM?** Real LLM calls are slow, expensive, and non-deterministic. Mock the LLM in unit tests so they run in milliseconds and produce consistent results. Only call the real LLM in smoke tests or integration tests.

### What this cell does
Defines `require_file()` (same as Part 03 — reused here for testing) and three test functions:

- **`test_require_file_missing()`** — calls `require_file()` with a path that doesn't exist, expects `FileNotFoundError`
- **`test_require_file_empty()`** — creates an empty file, calls `require_file()`, expects `ValueError`
- **`smoke_test_pipeline()`** — runs a minimal end-to-end check: creates the output directory and writes placeholder `report.json` and `report.md`

**How to read these tests:**
- Each test creates its own isolated temp directory (`tmp_path`) — no shared state between tests
- Tests assert the *type* of exception, not the message — so you can improve error messages without breaking tests
- The smoke test verifies the output contract: "after running the pipeline, these files must exist"

**Running with pytest:** Save these as `test_pipeline.py` and run `pytest test_pipeline.py -v`. Each function starting with `test_` is automatically discovered and run.

## Minimal test plan (3+ cases)

You should have at least:

- **happy path**: normal input works
- **edge case**: missing values or tiny CSV
- **failure case**: missing file / invalid schema

Practical tip: smoke tests can be “one command” checks that run in CI later. The key is making them deterministic enough to be repeatable.

```python
import json
import tempfile
from pathlib import Path


def require_file(path: str) -> Path:
    p = Path(path).expanduser()
    if not p.exists():
        raise FileNotFoundError("Input file not found: %s" % p)
    if p.stat().st_size == 0:
        raise ValueError("Input file is empty: %s" % p)
    return p


def test_require_file_missing(tmp_path):
    # TODO: implement pytest unit test for missing file.
    # Example: call require_file() and assert FileNotFoundError.
    missing = Path(tmp_path) / "missing.csv"
    try:
        _ = require_file(str(missing))
    except FileNotFoundError:
        return
    raise AssertionError("expected FileNotFoundError")


def test_require_file_empty(tmp_path):
    # TODO: implement pytest unit test for empty file.
    p = Path(tmp_path) / "empty.csv"
    p.write_text("", encoding="utf-8")
    try:
        _ = require_file(str(p))
    except ValueError:
        return
    raise AssertionError("expected ValueError")


def smoke_test_pipeline() -> None:
    # TODO: run an end-to-end pipeline with a tiny input.
    # Verify artifacts exist: output/report.json, output/report.md
    out_dir = Path("output")
    out_dir.mkdir(exist_ok=True)
    (out_dir / "report.json").write_text(json.dumps({"ok": True}, indent=2), encoding="utf-8")
    (out_dir / "report.md").write_text("# Report\n\nOK", encoding="utf-8")


with tempfile.TemporaryDirectory() as d:
    test_require_file_missing(d)
    test_require_file_empty(d)

smoke_test_pipeline()
print("wrote:", Path("output") / "report.json")
print("Implement pytest tests + smoke_test_pipeline().")
```

## Appendix: Solutions (peek only after trying)

Reference implementations for pytest-style tests, smoke test, and gitignore snippet.

```python
from pathlib import Path
def test_require_file_missing(tmp_path):
    import pytest

    missing = Path(tmp_path) / "missing.csv"
    with pytest.raises(FileNotFoundError):
        _ = require_file(str(missing))


def test_require_file_empty(tmp_path):
    import pytest

    p = Path(tmp_path) / "empty.csv"
    p.write_text("", encoding="utf-8")
    with pytest.raises(ValueError):
        _ = require_file(str(p))


def smoke_test_pipeline() -> None:
    out_dir = Path("output")
    out_dir.mkdir(exist_ok=True)
    (out_dir / "report.json").write_text(json.dumps({"ok": True}, indent=2), encoding="utf-8")
    (out_dir / "report.md").write_text("# Report\n\nOK", encoding="utf-8")


print("solution smoke wrote:", Path("output") / "report.json")
```

```python
def gitignore_snippet_todo() -> str:
    lines = [
        ".env",
        "output/",
        "__pycache__/",
        "*.pyc",
        ".pytest_cache/",
    ]
    return "\n".join(lines) + "\n"


snippet = gitignore_snippet_todo()
(OUTPUT_DIR / "gitignore_snippet_solution.txt").write_text(snippet, encoding="utf-8")
print("wrote:", OUTPUT_DIR / "gitignore_snippet_solution.txt")
```
